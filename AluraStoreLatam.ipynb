#1. Análisis de facturación

Ingreso total por cada tienda.

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos

Ingresos Totales Tiendas



```python
facturacion_tienda_1 = tienda_1['Precio'].sum()
print(f'Facturación tienda 1 : ${facturacion_tienda_1:.2f}')
facturacion_tienda_2 = tienda_2['Precio'].sum()
print(f'Facturación tienda 2 : ${facturacion_tienda_2:.2f}')
facturacion_tienda_3 = tienda_3['Precio'].sum()
print(f'Facturación tienda 3 : ${facturacion_tienda_3:.2f}')
facturacion_tienda_4 = tienda_4['Precio'].sum()
print(f'Facturación tienda 4 : ${facturacion_tienda_4:.2f}')

if (facturacion_tienda_1 > facturacion_tienda_2 and facturacion_tienda_1 > facturacion_tienda_3 and facturacion_tienda_1 > facturacion_tienda_4):
 print(f'La tienda con mayor ingresos es la tienda 1, con un total de:  ${facturacion_tienda_1:.2f}')

facturacion_total = facturacion_tienda_1 + facturacion_tienda_2 + facturacion_tienda_3 + facturacion_tienda_4
print(f'\nLa facturacion total de las 4 tiendas es : ${facturacion_total:.2f}')
```

```python
import matplotlib.pyplot as plt
#Datos de facturación
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [facturacion_tienda_1, facturacion_tienda_2, facturacion_tienda_3, facturacion_tienda_4]
colores = ['red', 'blue', 'green', 'purple']

#creando gráfico de barras
plt.figure(figsize=(8, 6))#tamaño gráfico
plt.bar(tiendas, ingresos, color=colores, label=tiendas)

#agregando valores sobre cada barra
for i, ingreso in enumerate(ingresos):
    plt.text(i, ingreso, f'${ingreso:.2f}', ha='center', va='bottom')
#ejes gráfico
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.title('Ingresos Totales por tiendas')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='lower right') # Generar la leyenda automáticamente
plt.show()
```

```python
# Calcular la distribución de ingresos
ingresos_totales_por_tienda = pd.Series([facturacion_tienda_1, facturacion_tienda_2, facturacion_tienda_3, facturacion_tienda_4], index=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
colores = ['red', 'blue', 'green', 'purple']
distribucion_ingresos = ingresos_totales_por_tienda / ingresos_totales_por_tienda.sum()

plt.figure(figsize=(7,7))
distribucion_ingresos.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=colores)
plt.title('Participación de Cada Tienda en el Total de Ingresos')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Función para calcular y mostrar las ventas por categoría de una tienda
def ventas_por_categoria(tienda, numero):
    ventas = tienda.groupby('Categoría del Producto')['Precio'].sum()
    print(f"\nVentas por categoría en Tienda {numero}:")
    categorias_populares = ventas.sort_values(ascending=False)
    print(f"Categorías más populares en Tienda {numero}:")
    for categoria, ventas_total in categorias_populares.items():
        print(f"{categoria}: ${ventas_total:.2f}")
    print("\n")  # Espacio para separar resultados

# Aplicar la función para cada tienda
ventas_por_categoria(tienda_1, 1)
ventas_por_categoria(tienda_2, 2)
ventas_por_categoria(tienda_3, 3)
ventas_por_categoria(tienda_4, 4)
```

```python
import matplotlib.pyplot as plt

def plot_ventas_por_categoria_bar(tienda, numero):
    ventas = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    ventas.plot(kind='bar', color='skyblue')
    plt.title(f'Ventas por Categoría en Tienda {numero}')
    plt.xlabel('Categoría del Producto')
    plt.ylabel('Ingresos')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

plot_ventas_por_categoria_bar(tienda_1, 1)
plot_ventas_por_categoria_bar(tienda_2, 2)
plot_ventas_por_categoria_bar(tienda_3, 3)
plot_ventas_por_categoria_bar(tienda_4, 4)
```

# 3. Calificación promedio de la tienda


```python
calific_1 = tienda_1['Calificación'].mean()
calific_2 = tienda_2['Calificación'].mean()
calific_3 = tienda_3['Calificación'].mean()
calific_4 = tienda_4['Calificación'].mean()

print(f'La Calificación de la Tienda N°1 es: {calific_1:.2f} ')
print(f'La Calificación de la Tienda N°2 es: {calific_2:.2f} ')
print(f'La Calificación de la Tienda N°3 es: {calific_3:.2f} ')
print(f'La Calificación de la Tienda N°4 es: {calific_4:.2f} ')
```

```python
import matplotlib.pyplot as plt

# Datos de calificación de las tiendas
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones = [calific_1, calific_2, calific_3, calific_4]

# Calcular la calificación promedio general de todas las tiendas
calificacion_promedio_general = (calific_1 + calific_2 + calific_3 + calific_4) / len(tiendas)

# Definir el rango de valores en el eje Y ajustado
min_calif = min(calificaciones) - 0.2  # Reducir un poco por debajo del mínimo
max_calif = max(calificaciones) + 0.2  # Expandir por encima del máximo

# Crear gráfico de barras vertical con escala ajustada
plt.figure(figsize=(8, 5))
plt.bar(tiendas, calificaciones, color=['green', 'blue', 'orange', 'red'])
plt.ylabel("Calificación Promedio")
plt.title("Calificación Promedio por Tienda")
plt.xticks(rotation=45, ha='right')

# Ajustar límites del eje Y para mejorar la visibilidad de diferencias
plt.ylim(min_calif, max_calif)

# Agregar valores sobre las barras
for i, v in enumerate(calificaciones):
    plt.text(i, v + 0.05, f"{v:.2f} ", ha='center', va='bottom', fontsize=12, fontweight='bold')

# Agregar líneas de cuadrícula
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Agregar línea horizontal para el promedio general
plt.axhline(calificacion_promedio_general, color='gray', linestyle='dashed', linewidth=2, label=f'Promedio General: {calificacion_promedio_general:.2f}')
plt.legend() # Mostrar la leyenda para la línea de promedio

plt.show()
```

```python
import matplotlib.pyplot as plt

# Concatenar datos de todas las tiendas
df_tiendas = pd.concat([tienda_1.assign(Tienda='Tienda 1'),
                        tienda_2.assign(Tienda='Tienda 2'),
                        tienda_3.assign(Tienda='Tienda 3'),
                        tienda_4.assign(Tienda='Tienda 4')])

# Agrupar y calcular precios por tienda y producto
productos_ordenados = df_tiendas.groupby(['Tienda', 'Producto'])['Precio'].sum().reset_index()

# Extraer los 10 productos más caros por tienda
top_10_por_tienda = productos_ordenados.sort_values(['Tienda', 'Precio'], ascending=[True, False]).groupby('Tienda').head(10)

# Definir los colores para cada tienda
colores_tiendas = {'Tienda 1': 'red', 'Tienda 2': 'blue', 'Tienda 3': 'green', 'Tienda 4': 'purple'}

# Crear gráfico
plt.figure(figsize=(12, 6))
for tienda in top_10_por_tienda['Tienda'].unique():
    subset = top_10_por_tienda[top_10_por_tienda['Tienda'] == tienda]
    plt.barh(subset['Producto'], subset['Precio'], label=tienda, color=colores_tiendas[tienda])

plt.xlabel('Precio')
plt.title('Productos Más y Menos Vendidos')
plt.legend()
plt.gca().invert_yaxis()  # Invertir eje Y para mejor legibilidad
plt.show()
```

```python
#Categorias más vendidas por tienda

# Colores personalizados para categorías
colores_por_categoria = {
    "Electrónicos": "magenta",
    "Electrodomésticos": "green",
    "Muebles": "red",
    "Instrumentos musicales": "orange",
    "otros": "grey"
}

# Función para graficar varias tiendas en subgráficos con orden y colores personalizados
def graficar_tiendas(dataframes, nombres_tiendas):
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # 2 filas, 2 columnas
    fig.suptitle('Categorías mas vendidas por Tiendas', fontsize=14)

    for ax, df, nombre_tienda in zip(axes.flatten(), dataframes, nombres_tiendas):
        ventas_por_categoria = df.groupby('Categoría del Producto')['Precio'].sum()
        ventas_por_categoria = ventas_por_categoria.sort_values(ascending=False)  # Orden de mayor a menor

        categorias = ventas_por_categoria.index
        ventas = ventas_por_categoria.values
        colores = [colores_por_categoria.get(categoria, "gray") for categoria in categorias]  # Color según categoría

        ax.barh(categorias, ventas, color=colores)
        ax.set_title(nombre_tienda, fontsize=12)
        ax.set_xlabel('Ventas Totales')
        ax.set_ylabel('Categoría del Producto')
        ax.invert_yaxis()
        ax.grid(axis='x', linestyle='--', alpha=0.7)

        # Mostrar valores en las barras
        for i, v in enumerate(ventas):
            ax.text(v, i, f"{int(v):,}", va='center', ha='left')

    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajusta para evitar solapamientos
    plt.show()

# Lista de DataFrames y nombres de tiendas
dataframes = [tienda_1, tienda_2, tienda_3, tienda_4]
nombres_tiendas = ["Tienda N°1", "Tienda N°2", "Tienda N°3", "Tienda N°4"]

# Aplicar la función para graficar todas las tiendas
graficar_tiendas(dataframes, nombres_tiendas)
```

Los resultados de este análisis servirán como base cuantitativa para fundamentar la recomendación final.

## **Análisis de Resultados**

### **1. Ingresos Totales**

El análisis de facturación total por tienda revela diferencias significativas en el rendimiento:

*   **Tienda 1:** $1,150,880,400.00
*   **Tienda 2:** $1,116,343,500.00
*   **Tienda 3:** $1,098,019,600.00
*   **Tienda 4:** $1,038,375,700.00

La Tienda 1 presenta la mayor facturación, mientras que la Tienda 4 registra la menor, con una diferencia de aproximadamente $112.5 millones respecto a la Tienda 1. Esta disparidad podría atribuirse a factores como la ubicación geográfica, antigüedad o nivel de consolidación en el mercado.

### **2. Ventas por Categoría y Productos Vendidos**

Todas las tiendas comercializan las mismas categorías principales: Electrónicos, Electrodomésticos y Muebles. Si bien estas categorías lideran las ventas en todas las tiendas, la Tienda 4 muestra un comportamiento relativamente mejor en la categoría de Juguetes en comparación con su rendimiento general.

En cuanto a productos específicos, los más vendidos suelen ser artículos de Muebles y Electrónicos, destacando en las Tiendas 1 y 3. Por el contrario, los Libros se encuentran consistentemente entre los productos menos vendidos en todas las tiendas.

### **3. Calificación Promedio de la Tienda**

Las calificaciones promedio de los clientes son notablemente similares entre las tiendas, todas alrededor de 4.00:

*   **Tienda 1:** 3.98
*   **Tienda 2:** 4.04
*   **Tienda 3:** 4.05
*   **Tienda 4:** 4.00

A pesar de ser la tienda con mayores ingresos, la Tienda 1 presenta la calificación promedio ligeramente más baja. Esto podría sugerir desafíos en el mantenimiento de altos estándares de servicio al cliente a gran volumen. La Tienda 4, con menor volumen de operaciones, mantiene una calificación promedio comparable a la de las tiendas más grandes.

### **4. Costo de Envío Promedio**

Se observa una correlación entre el volumen de ventas y el costo de envío promedio:

*   **Tienda 1:** $26,018.61
*   **Tienda 2:** $25,216.24

La Tienda 1, con el mayor volumen de ventas, también registra el costo de envío promedio más alto. La Tienda 4, con menor actividad comercial, presenta el costo de envío promedio más bajo.

Considerando los indicadores analizados, se recomienda al Sr. Juan vender la **Tienda 4**. Esta recomendación se fundamenta en los siguientes puntos:

*   **Menor Rendimiento Financiero:** La Tienda 4 consistentemente genera los menores ingresos totales en comparación con las otras tres tiendas.
